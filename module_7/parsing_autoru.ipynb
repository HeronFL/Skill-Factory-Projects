```python
brands = ['VOLKSWAGEN', 'NISSAN','MERCEDES', 'TOYOTA', 'AUDI', 'MITSUBISHI','SKODA','VOLVO','HONDA','INFINITI','LEXUS', 'BMW']
```

```python
# Собираем ссылки для заданных брендов в файл auto.json
url = 'https://auto.ru/moskva/cars/'
cars_urls = {}
counter = 0
for car_brand in brands:
    brand_url = url + car_brand + '/used/?page='
    cars_urls[car_brand] = []
    print(car_brand, counter)
    for number in range(1, 100):
        response = requests.get(brand_url+str(number),
                                headers={'User-Agent': 'Mozilla/5.0'})
        # Теперь создадим объект BeautifulSoup, указывая html парсер
        page = BeautifulSoup(response.text, 'html.parser')
        # Получаем все элементы с тегом 'a' и классом 'Link ListingItemTitle-module__link'
        link_list = page.find_all(
            'a', class_='Link ListingItemTitle-module__link')
        if len(link_list) != 0:
            for link in link_list:
                cars_urls[car_brand].append(link['href'])
                counter += 1
        else:
            break

with open('auto.json', 'w') as fp:
    json.dump(cars_urls, fp)
```

```python
# Собираем данные
counter = 0
cars = []

with open('auto.json') as f:
    avto = json.load(f)

for brand, cars_urls in avto.items():
    print(brand)
    for url in cars_urls:
        time.sleep(0.3)
        url_dict = {}

        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        response.encoding = 'utf-8'
        if response.status_code == 200:
            page = BeautifulSoup(response.text, 'html.parser')
            if 'Этот автомобиль уже продан' in page.text:
                continue
            else:
                script = page.find_all('script', attrs={'type': 'application/ld+json'})[0]
                details = json.loads(script.text)
                if 'bodyType' in details:
                    url_dict['bodyType'] = details['bodyType']
                else:
                    url_dict['bodyType'] = np.nan
                if 'brand' in details:
                    url_dict['brand'] = details['brand']
                else:
                    url_dict['brand'] = np.nan
                url_dict['car_url'] = url
                if 'color' in details:
                    url_dict['color'] = details['color']
                else:
                    url_dict['color'] = np.nan
                url_dict['description'] = details['description']
                url_dict['engineDisplacement'] = details['vehicleEngine']['engineDisplacement']
                url_dict['enginePower'] = details['vehicleEngine']['enginePower']
                #url_dict['fuelType'] = page.find_all('a', class_='Link Link_color_black')[3].text
                url_dict['fuelType'] = details['vehicleEngine']['fuelType']
                url_dict['image'] = details['image']
                url_dict['mileage'] = int(page.find('li', class_='CardInfoRow CardInfoRow_kmAge').find_all(
                    'span', class_='CardInfoRow__cell')[1].text.replace(u'\xa0', u' ')[:-3].replace(' ', ''))
                url_dict['modelDate'] = page.find_all('a', class_='Link Link_color_black')[0].text
                # url_dict['model_info'] =
                url_dict['model_name'] = str(page.find('script', attrs={
                    'type': 'application/json', 'id': 'initial-state'})).partition('"model_info":{"code":"')[2].partition('","name":"')[0]
                url_dict['name'] = details['name']
                if 'numberOfDoors' in details:
                    url_dict['numberOfDoors'] = details['numberOfDoors']
                else:
                    url_dict['numberOfDoors'] = np.nan
                url_dict['priceCurrency'] = details['offers']['priceCurrency']
                url_dict['price'] = details['offers']['price']
                url_dict['productionDate'] = details['productionDate']
                url_dict['vehicleConfiguration'] = details['vehicleConfiguration']
                if 'vehicleTransmission' in details:
                    url_dict['vehicleTransmission'] = details['vehicleTransmission']
                else:
                    url_dict['vehicleTransmission'] = np.nan 
                url_dict['Владельцы'] = page.find('li', class_='CardInfoRow CardInfoRow_ownersCount').find_all(
                    'span', class_='CardInfoRow__cell')[1].text.replace(u'\xa0', u' ')
                url_dict['ПТС'] = page.find('li', class_='CardInfoRow CardInfoRow_pts').find_all(
                    'span', class_='CardInfoRow__cell')[1].text
                url_dict['Привод'] = page.find('li', class_='CardInfoRow CardInfoRow_drive').find_all(
                    'span', class_='CardInfoRow__cell')[1].text
                url_dict['Руль'] = page.find('li', class_='CardInfoRow CardInfoRow_wheel').find_all(
                    'span', class_='CardInfoRow__cell')[1].text
                url_dict['Состояние'] = page.find('li', class_='CardInfoRow CardInfoRow_state').find_all(
                    'span', class_='CardInfoRow__cell')[1].text
                url_dict['Таможня'] = page.find('li', class_='CardInfoRow CardInfoRow_customs').find_all(
                    'span', class_='CardInfoRow__cell')[1].text
                complectation = []
                for i in page.find_all('li', class_='ComplectationGroups__itemContentEl'):
                    complectation.append(i.text)
                url_dict['complectation_dict'] = complectation
                cars.append(url_dict)
                counter += 1
    print(brand, counter)
    counter = 0
    brand_file = brand + '.csv'
    with open(brand_file, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = fieldnames = ['bodyType', 'brand', 'car_url', 'color', 'complectation_dict', 'description', 'engineDisplacement', 'enginePower',
                                   'fuelType', 'image', 'mileage', 'modelDate', 'model_name', 'name', 'numberOfDoors', 'parsing_unixtime', 'price', 'priceCurrency',
                                   'productionDate', 'vehicleConfiguration', 'vehicleTransmission', 'Владельцы', 'ПТС', 'Привод', 'Руль', 'Состояние', 'Таможня']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(cars)
print('end')
```
